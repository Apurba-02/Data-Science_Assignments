Matplotlib is a popular Python library used for data visualization. It provides a flexible and comprehensive set of tools for creating a wide range of static, interactive, and animated plots. Matplotlib is often used in data analysis, scientific research, and other fields where visualizing data is essential for understanding patterns, trends, and relationships.

The Pyplot module is a sub-library within Matplotlib that provides a simple interface for creating and customizing various types of plots. It makes it easier for users to create plots quickly without having to deal with the low-level details of Matplotlib's object-oriented interface.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1) Line Plot: A basic plot that displays data points connected by straight lines. It is used to visualize trends and patterns over time or continuous data.

2) Scatter Plot: This plot displays individual data points as dots on a 2D plane. It is used to examine the relationship between two continuous variables.

3) Bar Plot: A plot where rectangular bars are used to represent data. It is commonly used to compare categories or discrete data.

4) Histogram: A graphical representation of the distribution of numerical data. It shows the frequency of data within specified bins or intervals.

5) Pie Chart: A circular chart divided into sectors, each representing a proportion of the whole. It is used to show the relative sizes of different categories in a dataset.

Scatter Plot: This plot displays individual data points as dots on a 2D plane. It is used to examine the relationship between two continuous variables.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-Axis---->")
plt.ylabel("Y-Axis---->")
plt.title("Scatter Plot")
plt.show()
```

The subplot() function in Matplotlib is used to create multiple axes (plots) within a single figure. It allows you to arrange multiple subplots in a grid-like fashion, making it easier to visualize and compare different data visualizations side by side. The subplot() function is particularly useful when you want to display related plots or subplots together, providing a concise way to present complex visualizations.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 =  np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset along with key summary statistics. It visually displays the median, quartiles, and potential outliers of the data. The plot consists of a rectangular box (the "box") and two "whiskers" that extend from the box.

Box plots are used for several purposes:

1) Visualizing Data Distribution: Box plots give a clear visual representation of the spread and skewness of a dataset. They are particularly useful for identifying the presence of outliers.

2) Comparing Distributions: Box plots are excellent for comparing the distributions of multiple datasets side by side. They make it easy to see differences in medians, quartiles, and the spread of data.

3) Identifying Outliers: Outliers are easily identifiable as individual data points outside the whiskers, allowing for the detection of extreme values in the dataset.

4) Summarizing Data: Box plots provide a concise summary of essential statistics (median, quartiles) without displaying the raw data, making them ideal for presenting data in reports and presentations.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.show()

plt.boxplot(box2)
plt.show()
```
