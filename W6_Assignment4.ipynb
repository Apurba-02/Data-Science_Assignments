In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views in your web application. It allows you to define different routes that correspond to different pages or actions within your application.

Routes are defined using decorators in Flask, specifically the @app.route decorator. This decorator is used to associate a URL pattern with a function that will be executed when that URL is accessed. 

```python
# Q4
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def greet():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def info():
    return "<h3>Company Name: ABC Corporation</h3> <h3>Location: India</h3> <h3>Contact Detail: 999-999-9999</h3>"

if __name__ ==("__main__"):
    app.run(host="0.0.0.0")
```
