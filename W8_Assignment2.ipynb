```python
'''(1) pd.read_html() : It is a convenient function provided by Pandas to read HTML tables directly into a list 
#of DataFrame objects. It is particularly useful when you want to extract tabular data from HTML web pages 
and quickly convert it into DataFrame format for further analysis.'''

data = pd.read_html("https://www.basketball-reference.com/leagues/NBA_2015_totals.html")
df = data[0]
df
```

```python
# (2) DataFrame.head(n): Returns the first n rows of the DataFrame.

df.head(10)
```

```python
# (3) DataFrame.describe(): Generates descriptive statistics of numerical columns in the DataFrame.

df.describe()
```

```python
''' (4) DataFrame.info(): Provides information about the DataFrame, including the data types of each column 
and the number of non-null values. This helps in understanding the dataset's size and data types. '''

df.info()
```

```python
''' (5) DataFrame.groupby(): This function is used for grouping data based on one or more columns. 
It allows you to split the data into groups and then perform aggregate functions on each group. '''

g = df.groupby('Pos')
g.max()
```

```python
def add_weekday_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df

df8=pd.DataFrame({'Date': pd.date_range(start = '2023-01-01' , end = '2023-01-20')})
df8=add_weekday_column(df8)
df8
```

```python
def select_rows(df):
    df['Date'] = pd.to_datetime(df['Date'])
    selected_rows =pd.DataFrame( df[(df['Date'] >= '2023-01-01') & (df['Date'] <='2023-01-31')])
    return selected_rows
    
df9 = pd.DataFrame({'Date':pd.date_range(start= '2022-12-01', end = '2023-02-28')})
df10 = select_rows(df9)
df10
```
